```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
file_path1 = 'C:/Users/souvi/OneDrive/Desktop/Data Temperature/PC1/1 feb.xlsx'
file_path2 = 'C:/Users/souvi/OneDrive/Desktop/Data Temperature/PC1/2 feb.xlsx'
# file_path2 = 'C:/Users/souvi/OneDrive/Desktop/New folder (6)/PC2'
# Replace 'your_file.xlsx' with the actual path to your Excel file
df1 = pd.read_excel(file_path1)
# Left side
df2 = pd.read_excel(file_path2)
```

```python
# df1 = df1.iloc[16000:]
```

```python
# Assuming 'time_df' is your DataFrame with the time index and temperature data
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.scatter(df1.index, df1['Temperature'], marker='o', linestyle='-', color='b')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperature vs. Time')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.hist(df1['Temperature'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Histogram')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
plt.boxplot(df1['Temperature'], vert=False, notch=True, patch_artist=True)
plt.xlabel('Temperature')
plt.title('Temperature Box Plot')
plt.grid(True)
plt.show()
```

```python
# Calculate the mean, median, and mode of the 'temperature' column
mean_temp = df1['Temperature'].mean()
median_temp = df1['Temperature'].median()
mode_temp = df1['Temperature'].mode().iloc[0]  # In case of multiple modes
std_dev_temp = df1['Temperature'].std()

print(f"Mean Temperature: {mean_temp:.2f}")
print(f"Median Temperature: {median_temp:.2f}")
print(f"Mode Temperature: {mode_temp:.2f}")
print(f"Standard Deviation: {std_dev_temp:.2f}")
```

```python
def plot_function(df1, date):
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    # Scatter Plot
    axes[0].scatter(df1.index, df1['Temperature'], marker='o', linestyle='-', color='b')
    axes[0].set_xlabel('Time')
    axes[0].set_ylabel('Temperature')
    axes[0].set_title('Temperature vs. Time')
    axes[0].grid(True)
    
    # Histogram
    axes[1].hist(df1['Temperature'], bins=20, color='skyblue', edgecolor='black')
    axes[1].set_xlabel('Temperature')
    axes[1].set_ylabel('Frequency')
    axes[1].set_title('Temperature Histogram' +  date)
    axes[1].grid(True)
    
    # Adjust layout to prevent overlapping
    plt.tight_layout()
    
    # Display the combined plot
    plt.show()

def plot_function2(df1,date):
    # Scatter Plot
    plt.figure(figsize=(10, 6))
    plt.scatter(df1.index, df1['Temperature'], marker='o', linestyle='-', color='b')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.title('Temperature vs. Time : Date = ' + date)
    plt.grid(True)
    plt.show()
    
    # Histogram
    plt.figure(figsize=(10, 6))
    plt.hist(df1['Temperature'], bins=20, color='skyblue', edgecolor='black')
    plt.xlabel('Temperature')
    plt.ylabel('Frequency')
    plt.title('Temperature Histogram')
    plt.grid(True)
    plt.show()
```

```python
plot_function2(df2, '23.04.24')
```
